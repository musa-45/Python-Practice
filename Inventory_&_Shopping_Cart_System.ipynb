```python
# a) Inventory Dictionary
inventory = {
    "Apple": {"price": 50.0, "category": "food"},
    "Rice": {"price": 120.0, "category": "food"},
    "Headphones": {"price": 2500.0, "category": "electronics"},
    "Charger": {"price": 800.0, "category": "electronics"},
    "T-Shirt": {"price": 600.0, "category": "clothing"},
    "Jeans": {"price": 1500.0, "category": "clothing"}
}

# Display inventory
print("--- Available Products ---")
for product, details in inventory.items():
    print(f"{product:<12} : ৳ {details['price']:>7.2f}   [{details['category']}]")
```

```python
categories = set()

for details in inventory.values():
    categories.add(details["category"])

print("\n--- Available Categories ---")
print(", ".join(categories))
```

```python
cart = []

while True:
    item = input("\nEnter product (or 'done' to finish): ")

    if item.lower() == "done":
        break

    if item in inventory:
        cart.append(item)
    else:
        print(f"⚠ '{item}' not found in inventory. Try again.")

print("\nCart:", cart)
```

```python
def calculate_total(cart, inventory):
    total = 0

    for item in cart:
        total += inventory[item]["price"]

    return total

total = calculate_total(cart, inventory)

print(f"\nTotal (before discount): ৳ {total:.2f}")
```

```python
discount_percent = 0

if total >= 5000:
    discount_percent = 15
elif total >= 2000:
    discount_percent = 10
elif total >= 1000:
    discount_percent = 5

discount_amount = total * (discount_percent / 100)
final_amount = total - discount_amount

print("\n--- Discount Applied ---")
print(f"Discount applied : {discount_percent}% off")
print(f"Discount amount  : ৳ {discount_amount:.2f}")
print(f"Final amount     : ৳ {final_amount:.2f}")
```

```python
cart_summary = {}

for item in cart:
    if item in cart_summary:
        cart_summary[item] += 1
    else:
        cart_summary[item] = 1

print("\nCart Summary:", cart_summary)
```

```python
sorted_items = sorted(
    inventory.items(),
    key=lambda x: x[1]["price"],
    reverse=True
)

top_3 = tuple(item[0] for item in sorted_items[:3])

print("\nTop 3 most expensive items:", top_3)
```
