```python
#a

name = input("Enter you name: ")
num_subject = int(input("Enter the number of subjects (1-5): "))

while num_subject < 1 or num_subject > 5:
    num_subject = int(input("Invalid! Enter between 1-5:  "))
print(f"\n Welcom, {name}\n" )
```

```python
#b
scores =[]
for i in range(num_subject):
  while True:
    try:
          score = float(input(f"Enter score for subject {i+1}: "))
          if 0 <= score <= 100:
                scores.append(score)
                break
          else:
                print("Score must be between 0 and 100.")
    except:
            print("Invalid input! Enter a number.")
```

```python
def calculate_stats(scores):
    total = sum(scores)
    average = total / len(scores)
    highest = max(scores)
    lowest = min(scores)
    return total, average, highest, lowest
```

```python
def assign_grade(avg):
    if avg >= 80:
        return 'A'
    elif avg >= 65:
        return 'B'
    elif avg >= 50:
        return 'C'
    elif avg >= 35:
        return 'D'
    else:
        return 'F'
```

```python
total, average, highest, lowest = calculate_stats(scores)
grade = assign_grade(average)
```

```python
passed_subjects = [s for s in scores if s >= 50]
```

```python
print("\n========== Grade Report ==========")
print(f"Student  : {name}")
print(f"Scores   : {scores}")
print(f"Total    : {total}")
print(f"Average  : {average:.2f}")
print(f"Highest  : {highest}  |  Lowest: {lowest}")
print(f"Grade    : {grade}")
print(f"Passed   : {len(passed_subjects)} out of {num_subject} subjects")
```
